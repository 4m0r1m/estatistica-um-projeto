## TendÃªncia do meio de pagamento

### Quais os meios de pagamento mais utilizados?

Meios de pagamento mais de forma geral

```python
ax = sns.barplot(data=df_order_payments['payment_type'].value_counts())

ax.set_ylabel('Quantidade')
ax.set_xlabel('Tipo de pagamento')
ax.set_title('DistribuiÃ§Ã£o dos meios de pagamento por quantidade')

for i in ax.containers:
    ax.bar_label(i)

plt.show()
```

Meios de pagamento com maiores gastos

```python
ax = sns.barplot(df_order_payments.groupby(['payment_type'])['payment_value'].sum().sort_values(ascending=False))

ax.set_ylabel('Valor (R$)')
ax.set_xlabel('Tipo de pagamento')
ax.set_title('DistribuiÃ§Ã£o dos meios de pagamento por valor gasto')
ax.ticklabel_format(style='plain', axis='y')

plt.show()
```

Seguindo nas visualizaÃ§Ãµes acima, criamos um grÃ¡fico que mostra a distribuiÃ§Ã£o percentual de cada um destes itens:

```python
df_agg = df_order_payments.groupby(['payment_type'])['payment_value'].aggregate(['sum', 'count']).sort_values(by='count', ascending=False)

df_agg['valor'] = round(df_agg['sum'] / df_agg['sum'].sum() * 100, 3)
df_agg['quantidade'] = round(df_agg['count'] / df_agg['count'].sum() * 100, 3)

df_agg.reset_index(inplace=True)

df_melted = df_agg.melt(id_vars='payment_type', value_vars=['quantidade', 'valor'],
                        var_name='MÃ©trica', value_name='Porcentagem')

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='payment_type', y='Porcentagem', hue='MÃ©trica', data=df_melted)

ax.set_ylabel('Porcentagem (%)')
ax.set_xlabel('Tipo de pagamento')
ax.set_title('DistribuiÃ§Ã£o percentual dos meios de pagamento por quantidade e valor')

for i in ax.containers:
    ax.bar_label(i)

plt.show()
```

Observa-se que o percentual obtido em ambos os casos sÃ£o prÃ³ximos, indicando uma relaÃ§Ã£o direta entre quantidade de usos e o valor total das compras.

AlÃ©m disso, tanto via quantidade quanto pelo valor dos pedidos, o cartÃ£o de crÃ©dito fica bem na frente do que outros meios de pagamento, representando mais de 70% dos dados.

Vimos que o cartÃ£o de crÃ©dito Ã© amplamente utilizado, estando presente na maioria das compras e tambÃ©m sendo responsÃ¡vel pela maior quantidade de dinheiro gasto.

Isso significaria que o cartÃ£o Ã© o principal meio de pagamento em todas as faixas de preÃ§os? Ou ele seria apenas utilizado para produtos mais caros?

### Quais os meios de pagamento mais utilizados por faixa de preÃ§os?

Primeiro, vamos criar as faixas de preÃ§os:

```python
df_order_payments['payment_value'].describe()
```

```python
df_order_payments.query("payment_value == 0")
```

```python
sns.histplot(df_order_payments['payment_value'], bins=50, kde=True)
plt.title('DistribuiÃ§Ã£o de payment_value')
plt.xlabel('Valor de pagamento')
plt.ylabel('FrequÃªncia')
plt.show()
```

```python
bins = [0, 50, 100, 200, 500, 1000, df_order_payments['payment_value'].max()]
labels = ['0-50', '50-100', '100-200', '200-500', '500-1000', '1000+']

df_order_payments['faixa_pagamento'] = pd.cut(df_order_payments['payment_value'], bins=bins, labels=labels, include_lowest=True)
df_order_payments['faixa_pagamento'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='faixa_pagamento', hue='payment_type', data=df_order_payments)

ax.set_ylabel('Quantidade de transaÃ§Ãµes')
ax.set_xlabel('Faixa de valor de pagamento (R$)')
ax.set_title('DistribuiÃ§Ã£o de transaÃ§Ãµes por tipo de pagamento e faixa de valor')

plt.xticks(rotation=45)
plt.show()
```

```python
df_grouped = df_order_payments.groupby(['faixa_pagamento', 'payment_type']).size().reset_index(name='count')
df_grouped['percent'] = df_grouped.groupby('faixa_pagamento')['count'].transform(lambda x: x / x.sum() * 100)
df_grouped.sort_values(by=['faixa_pagamento', 'percent'], ascending=[True, False], inplace=True)
df_grouped #.query('faixa_pagamento == "1000+"')
```

```python
plt.figure(figsize=(12, 8))
ax = sns.barplot(x='faixa_pagamento', y='percent', hue='payment_type', data=df_grouped)

ax.set_ylabel('Percentual de transaÃ§Ãµes')
ax.set_xlabel('Faixa de valor de pagamento (R$)')
ax.set_title('DistribuiÃ§Ã£o de transaÃ§Ãµes por tipo de pagamento e faixa de valor')

plt.xticks(rotation=45)
plt.show()
```

```python
df_grouped['percent_cum'] = df_grouped.groupby('faixa_pagamento')['percent'].cumsum()

plt.figure(figsize=(12, 8))

for payment_type in df_grouped['payment_type'].unique():
    subset = df_grouped[df_grouped['payment_type'] == payment_type]
    plt.bar(subset['faixa_pagamento'], subset['percent'], bottom=subset['percent_cum'] - subset['percent'], label=payment_type)

for payment_type in df_grouped['payment_type'].unique():
    subset = df_grouped[df_grouped['payment_type'] == payment_type]
    for i in range(len(subset)):
        percent_value = subset['percent'].iloc[i]
        if percent_value >= 2:
            plt.text(
                subset['faixa_pagamento'].iloc[i],
                subset['percent_cum'].iloc[i] - subset['percent'].iloc[i] / 2,  # Centralizar na barra
                f"{percent_value:.2f}%",
                ha='center', va='center', color='white'
            )

plt.ylabel('Percentual de transaÃ§Ãµes')
plt.xlabel('Faixa de valor de pagamento (R$)')
plt.title('DistribuiÃ§Ã£o de transaÃ§Ãµes por tipo de pagamento e faixa de valor (100% Empilhado)')
plt.xticks(rotation=45)
plt.legend(title='Tipo de pagamento')
plt.tight_layout()

plt.show()
```

Algumas conclusÃµes sobre a anÃ¡lise dos meios de pagamento:

- Sobre o cartÃ£o de crÃ©dito:
    - Ele domina todas as faixas, mas aparece menos em transaÃ§Ãµes de valores mais baixos
    - Isso indicaria que as pessoas preferem utilizar menos o crÃ©dito em valores menores, dando preferÃªncia a outros meios
- Sobre o boleto:
    - Possui uma proporÃ§Ã£o prÃ³xima em todas as faixas
    - Aparece menos para valores mais altos, talvez indicando uma certa dificuldade do brasileiro conseguir pagar compras mais caras no boleto
- Sobre o voucher:
    - Aparece bem nas faixas mais baixas, mas diminui bastante a partir de R$ 100,00.
    - Talvez isso indique uma dificuldade em acessar vouchers para compras mais caras, pois geralmente vouchers seguem regras
- Sobre o debit_card:
    - Aparece pouco na base, nunca representando mais do que 2% de uma faixa
    - Similar ao boleto, aparece mais nas faixas mais baixas e diminui nas mais altas
    - Nas transaÃ§Ãµes acima de R$ 1000,00, o cartÃ£o de dÃ©bito aparece pela primeira vez maior do que o voucher

## TendÃªncia sobre os pedidos

### Como Ã© a distribuiÃ§Ã£o de pedidos por data?

Aqui Ã© interessante analisar se hÃ¡ um aumento de vendas/pedidos em certos perÃ­odos.

Para a nossa anÃ¡lise Ã© importante filtrar os pedidos, evitando aqueles que foram cancelados ou estÃ£o indisponÃ­veis.

```python
df_orders['order_status'].value_counts()
```

```python
df_orders = df_orders.query("order_status not in ['canceled', 'unavailable']")
df_orders
```

```python
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])
df_orders['day_order_purchase_timestamp'] = df_orders['order_purchase_timestamp'].dt.normalize()
df_day = df_orders.groupby(['day_order_purchase_timestamp']).size().reset_index(name='count')
df_day.sort_values(by='count', ascending=False).head(10)
```

```python
plt.figure(figsize=(16, 8))
sns.lineplot(x='day_order_purchase_timestamp', y='count', data=df_day, label='Contagem de Pedidos', marker='')

plt.xlabel('Data')
plt.ylabel('Contagem de Pedidos')
plt.title('TendÃªncia de Pedidos ao Longo do Tempo (DiÃ¡rio)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

> Houve um pico nas compras em e-commerces brasileiros no dia 24/11/2017 (Black Friday)

```python
df_orders['month'] = df_orders['order_purchase_timestamp'].dt.to_period('M')

df_monthly = df_orders.groupby('month').agg({'order_purchase_timestamp': 'first', 'order_id': 'size'}).reset_index()
df_monthly['order_purchase_timestamp'] = pd.to_datetime(df_monthly['order_purchase_timestamp']).dt.normalize()
df_monthly.rename(columns={'order_id': 'count'}, inplace=True)
df_monthly.sample(5)
```

```python
df_monthly['moving_avg'] = df_monthly['count'].rolling(window=3).mean()

plt.figure(figsize=(12, 6))
sns.lineplot(x='order_purchase_timestamp', y='count', data=df_monthly, label='Contagem de Pedidos', marker='o')
sns.lineplot(x='order_purchase_timestamp', y='moving_avg', data=df_monthly, color='orange',
             label='MÃ©dia MÃ³vel (3 meses)', linewidth=2)

for x, y in zip(df_monthly['order_purchase_timestamp'], df_monthly['count']):
    plt.text(x, y+50, f'{int(y)}', ha='center', va='bottom', fontsize=10, color='black')

plt.xlabel('Data')
plt.ylabel('Contagem de Pedidos')
plt.title('TendÃªncia de Pedidos ao Longo do Tempo (Mensal)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
num_months = 6
df_monthly['month_change'] = df_monthly['count'].pct_change().fillna(0)
print('Meses com uma alta diferenÃ§a quando comparados ao mÃªs passado (com outliers):')
df_monthly.nlargest(n=num_months, columns='month_change')
```

Aplicando remoÃ§Ã£o de outliers com IQR para identificar os meses com uma maior variaÃ§Ã£o em relaÃ§Ã£o ao seu mÃªs passado:

```python
Q1 = df_monthly["month_change"].quantile(0.25)
Q3 = df_monthly["month_change"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print('Meses com uma alta diferenÃ§a quando comparados ao mÃªs passado (sem outliers):')

highest_month_changes = df_monthly[
    (df_monthly["month_change"] >= lower_bound)
    & (df_monthly["month_change"] <= upper_bound)
].nlargest(n=num_months, columns="month_change")
highest_month_changes
```

Analisando os valores dos grÃ¡ficos e tabelas acima, aparentemente o dataset nÃ£o possui muitos valores de 2016 e do final de 2018, tornando-se perÃ­odos nÃ£o muito confiÃ¡veis para analisar. De qualquer forma, pode-se observar algumas coisas:

- A mÃ©dia mÃ³vel conseguiu seguir de forma boa o ritmo dos meses, indicando que a quantidade de pedidos prevista em 2018 provavelmente foi maior
- Com exceÃ§Ã£o do final de 2018 (provavelmente dados faltantes), a mÃ©dia do nÃºmero de pedidos foi ficando mais alta conforme o tempo, indicando, talvez, o sucesso e popularidade do e-commerce brasileiro que vemos nos anos atuais
- Em 2017, temos alguns momentos que chamam atenÃ§Ã£o, onde os valores aumentam bastante:
    - Em maio (3640 pedidos), comparado ao valor anterior de abril (2377 pedidos)
    - Em novembro (7423), comparado a outubro (4527)
    - Em janeiro (7187), comparado a dezembro (5620)
- Esses picos de pedidos podem ter sido causados por eventos sazonais, como:
    - Maio:
        - Dia das mÃ£es
    - Novembro - mÃªs com maior pico de vendas:
        - Black Friday
        - Cyber Monday (primeira segunda apÃ³s a Black Friday)
    - Janeiro:
        - LiquidaÃ§Ãµes de verÃ£o
        - Volta Ã s aulas

Com isso em mente, seria bom entender melhor como as vendas se comportam nesse perÃ­odo, verificando se hÃ¡ uma tendÃªncia no consumo dos brasileiros nestes no geral, mas tambÃ©m nestes perÃ­odos de pico.

### Quais as tendÃªncias de produtos nos pedidos?

Primeiro, precisamos juntar tudo em uma sÃ³ base, fazendo um merge entre:
- Pedidos (`df_orders`) - para buscar as informaÃ§Ãµes gerais do pedido
- Itens de cada pedido (`df_order_items`) - para buscar quais produtos estÃ£o no pedido
- Produtos (`df_products`) - para obter as categorias dos produtos do pedido

```python
orders_products = df_orders.merge(df_order_items, on='order_id', how='left').\
    merge(df_products[['product_id', 'product_category_name']], on='product_id', how='left')

orders_products
```

#### Como foi a distribuiÃ§Ã£o dos produtos geral e nos meses com picos de venda?

> NÃ£o faz muito sentido ter uma anÃ¡lise por **dia** sobre as categorias dos produtos, pois isso teria o objetivo de identificar se em algum dia houve uma maior venda de algum produto especÃ­fico e, para a nossa quantia de dados, uma anÃ¡lise por mÃªs se mostra mais eficiente

```python
orders_products['order_purchase_timestamp'] = pd.to_datetime(orders_products['order_purchase_timestamp'])
orders_products['month'] = orders_products['order_purchase_timestamp'].dt.to_period('M')
orders_products[['order_purchase_timestamp', 'month']].sort_values(by='month')
```

```python
plt.figure(figsize=(14, 6))

ax = sns.barplot(orders_products['product_category_name'].value_counts().nlargest(15).reset_index(),
            x='product_category_name', y='count')

ax.set_ylabel('NÃºmero de produtos')
ax.set_xlabel('Categoria do produto')
ax.set_title('Categorias que mais aparecem nos pedidos')

plt.xticks(rotation=90)
plt.show()
```

```python
monthly_products = orders_products.groupby(['month', 'product_category_name']).size().reset_index(name='count')
monthly_products
```

DistribuiÃ§Ã£o de produtos nos meses de maior mudanÃ§a

```python
num_categories = 10
peak_change_monthly_products = monthly_products[monthly_products['month'].isin(highest_month_changes['month'])] 
peak_change_monthly_products = (
    peak_change_monthly_products.groupby("month")
    .apply(lambda x: x.nlargest(num_categories, "count"))
    .reset_index(drop=True)
)
peak_change_monthly_products.sample(5)
```

```python
g = sns.FacetGrid(data=peak_change_monthly_products, col='month', col_wrap=2, height=6,
                  aspect=1.5, sharex=False)

g.map(sns.barplot, 'product_category_name', 'count')

g.set_axis_labels("Categoria de Produto", "Contagem")
g.figure.suptitle('Produtos mais comprados nos meses de maior variaÃ§Ã£o')
g.set_titles("MÃªs: {col_name}")
g.set_xticklabels(rotation=90)

plt.tight_layout()
plt.show()
```

```python
peak_months = df_monthly.sort_values(by='count', ascending=False).head(num_months)['month']
peak_monthly_products = monthly_products[monthly_products['month'].isin(peak_months)]
peak_monthly_products
```

```python
num_categories = 10
peak_monthly_products = (
    peak_monthly_products.groupby("month")
    .apply(lambda x: x.nlargest(num_categories, "count"))
    .reset_index(drop=True)
)
peak_monthly_products.sample(5)
```

```python
g = sns.FacetGrid(data=peak_monthly_products, col='month', col_wrap=2, height=6,
                  aspect=1.5, sharex=False)

g.map(sns.barplot, 'product_category_name', 'count')

g.set_axis_labels("Categoria de Produto", "Contagem")
g.figure.suptitle('Produtos mais comprados nos meses de pico')
g.set_titles("MÃªs: {col_name}")
g.set_xticklabels(rotation=90)

plt.tight_layout()
plt.show()
```
